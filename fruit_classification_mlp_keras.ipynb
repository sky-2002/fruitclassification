```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import os

fruit_class_training_folder = '/content/drive/MyDrive/Semester4/DS250/Training/Training_actual/Training/'
fruit_types = os.listdir('/content/drive/MyDrive/Semester4/DS250/Training/Training_actual/Training/')[:15]
# Initially trying on first 5 classes
print (fruit_types)  

print("Types of fruits found: ", len(fruit_types))
```

```python
fruits = []

for item in fruit_types:
 # Get all the file names
 all_fruits = os.listdir(fruit_class_training_folder + '/' +item)
 #print(all_shoes)

 # Add them to the list
 for fruit in all_fruits:
    fruits.append((item, str(fruit_class_training_folder + '/' +item) + '/' + fruit))
    #print(fruits[:1])
```

```python
# Build a dataframe        
fruits_df = pd.DataFrame(data=fruits, columns=['fruit type', 'image'])
print(fruits_df.head())
print(fruits_df.tail())
```

```python
import cv2
path = fruit_class_training_folder


#im_size = 100
im_size = 25

images = []
labels = []

for j,i in enumerate(fruit_types):
    data_path = path + str(i) 
    filenames = [i for i in os.listdir(data_path) ]
   
    for f in filenames:
        img = cv2.imread(data_path + '/' + f)  # reading that image as array
        #print(img)  # will get the image as an array
        img = cv2.resize(img, (im_size, im_size))
        images.append(img)
        labels.append(i)
        if (j+1)%1000==0:print(j+1,"images processed")
```

```python
# Transform the image array to a numpy type

images = np.array(images)

images.shape
```

```python
images = images.astype('float32') / 255.0
```

```python
images.shape
```

```python
from sklearn.preprocessing import LabelEncoder , OneHotEncoder

y=fruits_df['fruit type'].values
print(y[:15])
```

```python
# for y
y_labelencoder = LabelEncoder ()
y = y_labelencoder.fit_transform (y)
print (y)
print(len(y))
```

```python
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split

images, y = shuffle(images, y, random_state=1)

train_x, test_x, train_y, test_y = train_test_split(images, y, test_size=0.05, random_state=415)

#inpect the shape of the training and testing.
print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)
```

```python
# Develop a sequential model using tensorflow keras
import keras
import tensorflow as tf
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(25,25,3)),
    keras.layers.Dense(256, activation='relu'),
   
    keras.layers.Dense(5, activation=tf.nn.softmax)
])
```

```python
model.summary()
```

```python
# Compute the model parameters

model.compile(optimizer='adam', 
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Train the model  with 5 epochs 

model.fit(train_x,train_y, epochs=5)
```
