```python
def process_name(name):
  name = name.lower()
  names = name.split()
  return ("_").join(names)
```

```python
predictions = []
distances_euclid =[]
for i in range(len(test_images)):
    if i%1000==0:
      print(i,"images tested..")
    name = test_images.iloc[i]['image']
    image_id = ("_").join(name.lower().replace(" ","_").split("/"))
    pr,disteu = hog_test(image_id) 
    predictions.append(process_name(pr))
    distances_euclid.append(disteu)
```

```python
test_images['prediction'] = predictions
test_images['Euclidean distance'] = distances_euclid
```
